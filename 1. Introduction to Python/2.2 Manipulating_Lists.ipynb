List Manipulation
- Change list elements
- Add list elements
- Remove list elements


```python
x = ["a", "b", "c"]
y = x
y[1] = "z"
y
```

```python
x
```

If you want to create a list y that points to a new list in the memory with the same values,

```python
x = ["a", "b", "c"]
y = list(x)
y = x[:]
y[1] = "z"
x
```

```python
y
```

```python
# Create the areas list
areas = ["hallway", 11.25, "kitchen", 18.0, "living room", 20.0, "bedroom", 10.75, "bathroom", 9.50]

# Correct the bathroom area
areas[9] = 10.50

# Change "living room" to "chill zone"
areas[4] = "chill zone"

print(areas)
```

```python
# Create the areas list and make some changes
areas = ["hallway", 11.25, "kitchen", 18.0, "chill zone", 20.0,
         "bedroom", 10.75, "bathroom", 10.50]

# Add poolhouse data to areas, new list is areas_1
areas_1 = areas + ["poolhouse", 24.5]
print(areas_1)

# Add garage data to areas_1, new list is areas_2
areas_2 = areas_1 + ["garage", 15.45]
print(areas_2)
```

Pay attention here: as soon as you remove an element from a list, the indexes of the elements that come after the deleted element all change!

Presta atención: en cuanto eliminas un elemento de una lista, los índices de los elementos que vienen después del elemento eliminado cambian.

```python
areas = ["hallway", 11.25, "kitchen", 18.0,
        "chill zone", 20.0, "bedroom", 10.75,
         "bathroom", 10.50, "poolhouse", 24.5,
         "garage", 15.45]
del(areas[-4:-2])
print(areas)
```

```python
# Create list areas
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Create areas_copy
areas_copy = list(areas)

# Change areas_copy
areas_copy[0] = 5.0

# Print areas
print(areas)
```

Nice! The difference between explicit and reference-based copies is subtle, but can be really important. Try to keep in mind how a list is stored in the computer's memory
