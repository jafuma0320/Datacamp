```python
# Importing pandas and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("/content/drive/MyDrive/Datacamp/6. Project Investigating Netflix Movies/netflix_data.csv")

netflix_df.head(10)
```

```python
# 1. Filter the data for movies released in the 1990s

# You'll want to subset the DataFrame to keep only the movies
#and
#then filter the years so that you are working with rows where the release year is between 1990 and 1999.

# Subset the DataFrame for type "Movie"
movies_df = netflix_df[netflix_df["type"] == "Movie"]
```

```python
# Filter the to keep only movies released in the 1990s
movies_released_90s_df = movies_df [np.logical_and(movies_df["release_year"] >= 1990, movies_df["release_year"] < 1999)]
print(movies_released_90s_df[["type", "release_year"]].head(10))
```

```python
# Find the most frequent movie duration
# Visualize the data to view the distributions of durations for movies released in the 1990s.

# Visualizing a distribution

# Use plt.hist() on your desired column to view the distribution of its values.
plt.figure(figsize=(10, 6))
plt.hist(movies_released_90s_df['duration'], bins=20, edgecolor='black')
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()

# Look at the plot to see where the most frequent count falls and save the duration value as duration.

duration = 100
```
